## ‚öôÔ∏è Í∏∞Îä• 1-1. Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞ : data_load()
- csv Î™®ÎìàÎ°ú Îç∞Ïù¥ÌÑ∞Î•º Î∂àÎü¨Ïò¥
  - Ï≤´ rowÎäî column Ïù¥Î¶ÑÏúºÎ°ú Ïì∞Í∏∞ÏúÑÌï¥ Îî∞Î°ú Ï†ÄÏû•Ìï®
  - ÎÇòÎ®∏ÏßÄ rowÏùÄ dataÎ°ú Ïì∞Í∏∞ÏúÑÌï¥ Î¶¨Ïä§Ìä∏Î°ú Ï†ÄÏû•Ìï®
- Îç∞Ïù¥ÌÑ∞ Ï†ÑÏ≤òÎ¶¨Î•º ÏõêÌôúÌïòÍ≤å ÌïòÍ∏∞ÏúÑÌï¥ pandasÎ°ú DataFrameÏúºÎ°ú ÎßåÎì§Í≥† Ï∂úÎ†•Ìï®
  - Í∑∏ÎåÄÎ°ú ÎßåÎì§Í≤åÎêòÎ©¥ Îç∞Ïù¥ÌÑ∞Í∞Ä Î¨∏ÏûêÏó¥Î°ú Ï†ÄÏû•ÎêòÍ∏∞ ÎïåÎ¨∏Ïóê dtypeÎ≥ÄÌôòÎèÑ Ïã§ÏãúÌï®

image.png

```python
# Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞ Í∏∞Îä•
def data_load(filepath):
    '''
    ÏûÖÎ†•Í∞í Ï†ïÎ≥¥
    filepath : csv ÌååÏùº ÎîîÎ†âÌÜ†Î¶¨
    '''
    # open Ìï®ÏàòÎ°ú ÌååÏùºÏùÑ Ïó¥Í∏∞ (Ï≤òÎ¶¨Í∞Ä ÎÅùÎÇòÎ©¥ ÌååÏùºÏùÑ Îã´ÎèÑÎ°ù withÎ¨∏ ÌôúÏö©)
    with open(file=filepath) as csv_file:
        # csv Î™®ÎìàÎ°ú ÌååÏùºÏùÑ ÏùΩÍ∏∞ (reader)
        csv_reader = csv.reader(csv_file)
        # Ï≤´ Ïó¥ Îî∞Î°ú Ï†ÄÏû•
        row0 = next(csv_reader)
        # ÎëêÎ≤àÏß∏ Ïó¥Î∂ÄÌÑ∞ Îç∞Ïù¥ÌÑ∞Î•º listÎ°ú Ï†ÄÏû•
        data = list(csv_reader)
    
    # pandasÎ°ú DataFrame ÏÉùÏÑ± (column : row0)
    dataframe = pd.DataFrame(data=data, columns=row0)
    # ÎèÖÎ¶ΩÎ≥ÄÏàò(x1~x8)Îäî floatÎ°ú, Ï¢ÖÏÜçÎ≥ÄÏàò(y)Îäî intÎ°ú Î≥ÄÌôò
    dataframe = dataframe.astype(float).astype({'y':int})
    
    # Ï∂úÎ†• : DataFrame
    return dataframe
```

- Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞ Í∏∞Îä• ÎèôÏûë ÌÖåÏä§Ìä∏

```python
# data_load() Ïã§Ìñâ (filepath : csv fileÏùò directory)
df = data_load(filepath=csv_dir)

# dfÏùò columnÌôïÏù∏
print(f"List of columns : {list(df.columns)}")
# dfÏùò shapeÌôïÏù∏
print(f"DataFrame shape : {df.shape}")
```

- numpyÎ°ú Îí§ÏÑûÏùÑ Ïù∏Îç±Ïä§ Î≤àÌò∏ arrayÎ•º ÏÉùÏÑ±
  - ÎèôÏùºÌïú Í≤∞Í≥ºÍ∞Ä ÎÇòÏò§ÎèÑÎ°ù ÌïòÍ∏∞ ÏúÑÌï¥ random seedÎ•º ÏßÄÏ†ïÌï®(Í∏∞Î≥∏Í∞í=2023)
- pandasÎ•º ÌÜµÌï¥ DataFrameÏùò Ìñâ ÏàúÏÑúÎ•º Îí§ÏÑûÍ∏∞
  - .reindexÎ•º ÌÜµÌï¥ Ìñâ ÏàúÏÑúÎ•º arrayÏóê Îî∞Îùº Îí§ÏÑûÏùå
  - .reset_index ÏòµÏÖòÏùÑ Ï∂îÍ∞ÄÌï¥ Îí§ÏÑûÏùÄ Îç∞Ïù¥ÌÑ∞Ïùò Ïù∏Îç±Ïä§Î•º Ï¥àÍ∏∞ÌôîÌï†ÏßÄ Í≤∞Ï†ïÌïòÎèÑÎ°ù Ìï®

```python
# Îç∞Ïù¥ÌÑ∞ Îí§ÏÑûÍ∏∞ Í∏∞Îä•
def data_shuffle(data, reset_index=True, seed=2023):
    '''
    ÏûÖÎ†•Í∞í Ï†ïÎ≥¥
    data : Îí§ÏÑûÏùÑ Îç∞Ïù¥ÌÑ∞(DataFrame)
    reset_index : Í∏∞Î≥∏Í∞í True, Îí§ÏÑûÏù∏ indexÎ•º Ïû¨ÏÑ§Ï†ïÌï†ÏßÄ Ïó¨Î∂Ä
    seed : Í∏∞Î≥∏Í∞í 2023, numpy.random.seed Í∞í
    '''
    # Îç∞Ïù¥ÌÑ∞ Í∏∏Ïù¥Î•º data_lengthÎ≥ÄÏàòÏóê Ï†ÄÏû•
    data_length = data.shape[0]

    # index Ï¥àÍ∏∞Í∞í : 0 ~ n-1
    shuffle_idx = np.arange(data_length)
    # seed ÏÑ§Ï†ï (Í∏∞Î≥∏Í∞íÏùÄ 2023)
    np.random.seed(seed)
    # index array Îí§ÏÑûÍ∏∞
    np.random.shuffle(shuffle_idx)

    # indexÎ•º Ï¥àÍ∏∞ÌôîÌïòÏßÄ ÏïäÏùÑ Í≤ΩÏö∞
    if reset_index==False:
        # shuffle_idxÎ°ú reindexÎßå Ïã§Ïãú
        df = data.reindex(index=shuffle_idx)
    # indexÎ•º Ï¥àÍ∏∞ÌôîÌï† Í≤ΩÏö∞(Í∏∞Î≥∏Í∞íÏúºÎ°ú ÏÑ§Ï†ïÎêòÏñ¥ÏûàÏùå)
    elif reset_index==True:
        # shuffle_idxÎ°ú reindex Ïã§Ïãú + reset_indexÎèÑ Ïã§Ïãú
        df = data.reindex(index=shuffle_idx).reset_index(drop=True)
    
    # Ï∂úÎ†• : Îí§ÏÑûÏù∏ dataframe
    return df
```

- Îç∞Ïù¥ÌÑ∞ Îí§ÏÑûÍ∏∞ Í∏∞Îä• ÎèôÏûë ÌÖåÏä§Ìä∏

```python
# data_shuffle()Ïã§Ìñâ (reset_index=False)
df_sf0 = data_shuffle(df, reset_index=False)
# data_shuffle()Ïã§Ìñâ (reset_index=True)
df_sf = data_shuffle(df)

# Ìï®Ïàò Ïã§Ìñâ Í≤∞Í≥º Ï∂úÎ†•
print("Shuffled DataFrame (not reset index)")
display(df_sf0.head(10))
print("\nShuffled DataFrame (reset index)")
display(df_sf.head(10))
```

## ‚öôÔ∏è Í∏∞Îä• 2-2. Îç∞Ïù¥ÌÑ∞ ÌëúÏ§ÄÌôî Í∏∞Îä• : standard_scaler()
- Îç∞Ïù¥ÌÑ∞Ïùò featureÎì§Ïùò ÌèâÍ∑†Í≥º ÌëúÏ§ÄÌé∏Ï∞®Í∞Ä Ï†úÍ∞ÅÍ∞ÅÏù¥ÎØÄÎ°ú Îç∞Ïù¥ÌÑ∞Ïùò Î∂ÑÌè¨Î•º ÎßûÏ∂îÍ∏∞ ÏúÑÌï¥ ÌëúÏ§ÄÌôî Í∏∞Îä•ÏùÑ Ï∂îÍ∞ÄÏ†ÅÏúºÎ°ú Íµ¨ÌòÑÌï®
  - ÌëúÏ§ÄÌôî Î∞©ÏãùÏùÄ ÏùåÏàò Îç∞Ïù¥ÌÑ∞Îì§ÎèÑ Ìè¨Ìï®ÎêòÏñ¥ ÏûàÍ∏∞ ÎïåÎ¨∏Ïóê standard scaling Î∞©ÏãùÏúºÎ°ú ÌëúÏ§ÄÌôîÎ•º ÏßÑÌñâÌï® (ÌèâÍ∑†0, ÌëúÏ§ÄÌé∏Ï∞®1)
- pandasÎ•º Ïù¥Ïö©Ìïú columnÎã®ÏúÑÏùò ÌëúÏ§ÄÌôî ÏßÑÌñâ
  - X, yÎ•º Î®ºÏ†Ä Î∂ÑÎ¶¨ÌïúÎã§Ïùå, XÏóêÎåÄÌïú ÌëúÏ§ÄÌôî ÏßÑÌñâ ÌõÑ Îã§Ïãú yÏôÄ Î≥ëÌï©ÌïòÎäî Î∞©Ïãù

```python
# Îç∞Ïù¥ÌÑ∞ ÌëúÏ§ÄÌôî Í∏∞Îä•
def standard_scaler(data, output_dim):
    '''
    ÏûÖÎ†•Í∞í Ï†ïÎ≥¥
    data : ÌëúÏ§ÄÌôîÏãúÌÇ¨ Îç∞Ïù¥ÌÑ∞(DataFrame)
    output_dim : Ï∂úÎ†•Ï∏µÏùò ÎÖ∏Îìú Ïàò(Ï¢ÖÏÜçÎ≥ÄÏàò Ïàò)
    '''
    # XÏôÄ yÎ•º ÏùºÎã® Î∂ÑÎ¶¨ÏãúÌÇ§ÎèÑÎ°ù Ìï®
    X = data.iloc[:,:-output_dim]
    y = data.iloc[:,-output_dim:]

    # pandasÏó∞ÏÇ∞Ïóê ÏùòÌï¥ column Î≥ÑÎ°ú Ïó∞ÏÇ∞Ïù¥ ÏßÑÌñâÎêúÎã§(ÌèâÍ∑†Í∞íÏùÑ ÎπºÍ≥† ÌëúÏ§ÄÌé∏Ï∞®Î°ú ÎÇòÎàÑÍ∏∞)
    std_scale_X = (X-X.mean()) / X.std()

    # ÌëúÏ§ÄÌôîÌïú XÎ•º yÏôÄ concatÌïòÏó¨ Îã§Ïãú DataFrameÏùÑ Ï†ïÏùò
    df = pd.concat([std_scale_X, y], axis = 1)

    # Ï∂úÎ†• ÌëúÏ§ÄÌôîÎ•º ÎßàÏπú dataframe
    return df
```

- Îç∞Ïù¥ÌÑ∞ ÌëúÏ§ÄÌôî Í∏∞Îä• ÎèôÏûë ÌÖåÏä§Ìä∏

```python
# standard_scaler Ìï®Ïàò Ïã§Ìñâ
df_sc = standard_scaler(df_sf, output_dim=1)

# ÌëúÏ§ÄÌôîÎ•º Í±∞Ïπú dataframe ÌôïÏù∏
print(f"Scaled DataFrame shape : {df.shape}")
display(df_sc.head(10))
```

## ‚öôÔ∏è Í∏∞Îä• 2-3. ÌïôÏäµ Î∞è ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨ Í∏∞Îä• : train_test_split()
- pandasÎ•º ÌôúÏö©Ìïú ÌïôÏäµ Î∞è ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨ Í∏∞Îä•
  - ÌïôÏäµÎç∞Ïù¥ÌÑ∞ : ÌÖåÏä§Ìä∏Îç∞Ïù¥ÌÑ∞ = 0.8 : 0.2
  - ÎØ∏ÎãàÎ∞∞Ïπò ÏÇ¨Ïù¥Ï¶à : 4

```python
# ÌïôÏäµ Î∞è ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨ Í∏∞Îä•
def train_test_split(data, train_ratio, batch_size):
    '''
    ÏûÖÎ†•Í∞í Ï†ïÎ≥¥
    data : Î∂ÑÎ¶¨Ìï† Îç∞Ïù¥ÌÑ∞(DataFrame)
    trian_ratio : ÌõàÎ†® Îç∞Ïù¥ÌÑ∞ ÎπÑÏú® ( 0 ~ 1 )
    batch_size : mini-batch size (integer)
    '''
    # Îç∞Ïù¥ÌÑ∞Ïùò Ìñâ Í∞ØÏàòÎ•º data_sizeÎ°ú ÏßÄÏ†ï
    data_size = data.shape[0]
    # train_ratioÎ•º Í≥±Ìïú Í∞íÏùÑ Î∞∞ÏπòÏÇ¨Ïù¥Ï¶àÎ°ú ÎÇòÎàÑÏñ¥ batchÎç∞Ïù¥ÌÑ∞Ïùò Í∞ØÏàòÎ•º Íµ¨Ìï®
    batch_count = int(data_size*train_ratio) // batch_size
    
    # bacthÎç∞Ïù¥ÌÑ∞ Í∞ØÏàòÏôÄ ÏÇ¨Ïù¥Ï¶àÎ•º Í≥±Ìïú Í∞íÏù¥ test Îç∞Ïù¥ÌÑ∞Ïùò ÏãúÏûë Ïù∏Îç±Ïä§Í∞Ä Îê®
    test_1st_idx = batch_count*batch_size

    # train, test Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨(pandas)
    train = data.iloc[:test_1st_idx,:]
    test = data.iloc[test_1st_idx:,:]

    # Ï∂úÎ†• : train, test dataframe
    return train, test
```

- ÌïôÏäµ Î∞è ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨ Í∏∞Îä• ÎèôÏûë ÌÖåÏä§Ìä∏

```python
# train_test_split() Ïã§Ìñâ(ratioÎäî 0.8, batch_sizeÎäî 4)
train, test = train_test_split(df_sc, train_ratio=0.8, batch_size=4)

# train, test Ï∂úÎ†•Ìï¥Î≥¥Í∏∞
print(f"train dataset : {train.shape}")
display(train)
print(f"\ntest dataset : {test.shape}")
display(test)
```

## ‚öôÔ∏è Í∏∞Îä• 2-4. ÎèÖÎ¶ΩÎ≥ÄÏàò Î∞è Ï¢ÖÏÜçÎ≥ÄÏàò Î∂ÑÎ¶¨ Í∏∞Îä• : X_y_split()
- pandasÏôÄ numpyÎ•º ÌôúÏö©Ìï¥ ÎèÖÎ¶ΩÎ≥ÄÏàò(x1~x8)ÏôÄ Ï¢ÖÏÜçÎ≥ÄÏàò(y)Î°ú Î∂ÑÎ¶¨ÌïòÎäî Í∏∞Îä•
  - Ï∂úÎ†•Ï∏µ ÎÖ∏ÎìúÏàòÏóê Îî∞Îùº Îç∞Ïù¥ÌÑ∞Î•º X ÏôÄ yÎ°ú Î∂ÑÎ¶¨Ìï®
  - Ï∂úÎ†•ÏùÄ ÌñâÎ†¨ Ïó∞ÏÇ∞ÏùÑ ÏàòÏõîÌïòÍ≤å ÌïòÍ∏∞ ÏúÑÌï¥ ndarrayÎ°ú Ï∂úÎ†•Ìï®
- Î≥∏ Îç∞Ïù¥ÌÑ∞ÏóêÏÑ† yÍ∞Ä ÌïòÎÇòÎøêÏù¥Ïñ¥ÏÑú output_dimÏù¥ 1Ïù∏Îç∞, Ïù¥Îïå ÌòïÌÉúÎ•º (n, ) ÌòπÏùÄ (n, 1)Î°ú ÎßûÏ∂∞ Ï§Ñ Ïàò ÏûàÎã§. 
  - Í∑∏Îü∞Îç∞ (n,1)Î°ú ÎßûÏ∂∞Ï§òÏïºÎßå ÌñâÎ†¨ Ïó∞ÏÇ∞Ïóê Ïò§Î•òÍ∞Ä Î∞úÏÉùÌïòÏßÄ ÏïäÏúºÎØÄÎ°ú Ï£ºÏùò

```python
# ÎèÖÎ¶ΩÎ≥ÄÏàò Î∞è Ï¢ÖÏÜçÎ≥ÄÏàò Î∂ÑÎ¶¨
def X_y_split(data, output_dim):
    '''
    ÏûÖÎ†•Í∞í Ï†ïÎ≥¥
    data : Î∂ÑÎ¶¨Ìï† Îç∞Ïù¥ÌÑ∞ (DataFrame)
    output_dim : Ï∂úÎ†•Ï∏µ ÎÖ∏Îìú Ïàò (Ï¢ÖÏÜçÎ≥ÄÏàò Ïàò)
    '''
    # XÎäî Îí§ÏóêÏÑú Î∂ÄÌÑ∞ output_dimÏßÅÏ†ÑÍπåÏßÄÏùò column Îç∞Ïù¥ÌÑ∞
    X = np.array(data.iloc[:,:-output_dim])
    # yÎäî Îí§ÏóêÏÑúÎ∂ÄÌÑ∞ output_dimÏù¥ÌõÑÏùò columnÎç∞Ïù¥ÌÑ∞ (ÎÅùÏóê : Î•º Íº≠ Î∂ôÏó¨Ï£ºÏûê)
    y = np.array(data.iloc[:,-output_dim:])

    # Ï∂úÎ†• : X, y ndarray
    return X, y
```

- ÎèÖÎ¶ΩÎ≥ÄÏàò Î∞è Ï¢ÖÏÜçÎ≥ÄÏàò Î∂ÑÎ¶¨ Í∏∞Îä• ÎèôÏûë ÌÖåÏä§Ìä∏

```python
# X_y_split()Ìï®Ïàò Ïã§ÌñâÌïòÏó¨ X, y arrayÎ•º Ï†ÄÏû• (output_dim = 1)
X_train, y_train = X_y_split(train, output_dim=1)
X_test, y_test = X_y_split(test, output_dim=1)

# Î∂ÑÎ¶¨ Ï†Ñ Îç∞Ïù¥ÌÑ∞ shape Ï∂úÎ†•
print("Data Shape (Before)")
print(f"train : {train.shape}")
print(f"test : {test.shape}")

# Î∂ÑÎ¶¨ ÌõÑ Îç∞Ïù¥ÌÑ∞ shape Ï∂úÎ†•
print("\nData Shape (Splited)")
print(f"X_train : {X_train.shape}")
print(f"y_train : {y_train.shape}")
print(f"X_test : {X_test.shape}")
print(f"y_test : {y_test.shape}")
```

## ‚öôÔ∏è Í∏∞Îä• 3-1. Í∞ÄÏ§ëÏπò, Ìé∏Ìñ• ÏÉùÏÑ± Í∏∞Îä• : initialization_parameter()
- numpyÏùò random Î™®ÎìàÏùÑ ÌÜµÌï¥ Í∞ÄÏ§ëÏπò Î∞è Ìé∏Ìñ•Ïóê ÎåÄÌïú ÎÇúÏàòÎ•º ÏÉùÏÑ±
  - Î∂ÑÌè¨Î•º Îç∞Ïù¥ÌÑ∞ÏôÄ ÎßûÏ∂∞Ï£ºÍ∏∞ ÏúÑÌï¥ .randnÏúºÎ°ú ÌèâÍ∑†0, ÌëúÏ§ÄÌé∏Ï∞®1Ïù∏ ÎÇúÏàòÎì§Î°ú ÏÉùÏÑ±Ìï®
  - ÎèôÏùºÌïú Í≤∞Í≥ºÎ•º ÎÇº Ïàò ÏûàÎèÑÎ°ù random seedÎ•º ÏÑ§Ï†ï (Í∏∞Î≥∏Í∞íÏùÄ 2023)
- Ïù∏Í≥µÏã†Í≤ΩÎßùÏóêÏÑú initÌï®Ïàò (Ï¥àÍ∏∞Ìï®Ïàò)ÏóêÏÑú ÏÇ¨Ïö©Îê† ÏòàÏ†ï

```python
# Í∞ÄÏ§ëÏπò, Ìé∏Ìñ• ÏÉùÏÑ± Í∏∞Îä•
def initialization_parameter(input_dim, output_dim, seed=2023):
    '''
    ÏûÖÎ†•Í∞í Ï†ïÎ≥¥
    input_dim : ÏûÖÎ†•Ï∏µ ÎÖ∏ÎìúÏàò
    output_dim : Ï∂úÎ†•Ï∏µ ÎÖ∏ÎìúÏàò
    seed : Í∏∞Î≥∏Í∞íÏùÄ 2023, numpy.random.seedÍ∞í
    '''
    # seed ÏÑ§Ï†ï (Í∏∞Î≥∏Í∞íÏùÄ 2023ÏúºÎ°ú ÏÑ§Ï†ï)
    np.random.seed(seed)
    
    # weight ndarrayÏÉùÏÑ±, shape : (input_dim,output_dim)
    weights = np.random.randn(input_dim,output_dim)
    # bias ndarrayÏÉùÏÑ±, shape : (output_dim,)
    bias = np.random.randn(output_dim)

    # Ï∂úÎ†• : weights, bias
    return weights, bias
```

- Í∞ÄÏ§ëÏπò Î∞è Ìé∏Ìñ• ÏÉùÏÑ± Í∏∞Îä• ÎèôÏûë ÌÖåÏä§Ìä∏

```python
# initialization_parameter Ìï®Ïàò Ïã§Ìñâ
weights, bias = initialization_parameter(input_dim=8, output_dim=1)

# weight Ï∂úÎ†•Ìï¥Î≥¥Í∏∞
print(f"input dimension : {len(weights)}")
print(f"\nweights :\n{weights}")

# bias Ï∂úÎ†•Ìï¥Î≥¥Í∏∞
print(f"\noutput dimension : {len(bias)}")
print(f"\nbias : {bias}")
```

## üìä Í∏∞Îä• 3-2. Í∞ÄÏ§ëÏπò ÏãúÍ∞ÅÌôî Í∏∞Îä• : weights_plot()
- Ï∂úÎ†•Ï∏µÏù¥ 1Í∞úÎøêÏù¥ÎØÄÎ°ú Í∞ÄÏ§ëÏπòÎì§Ïóê ÎåÄÌïú ÏãúÍ∞ÅÌôîÌï† Ïàò ÏûàÏùÑ Í≤ÉÏù¥Îùº ÌåêÎã®Ìï¥ Í∏∞Îä•ÏùÑ Ï∂îÍ∞Ä Íµ¨ÌòÑÌïòÏòÄÏùå
  - Ïù∏Í≥µÏã†Í≤ΩÎßù classÏóêÎèÑ Ìè¨Ìï®ÏãúÏºúÏÑú ÌïôÏäµ Ï†ÑÌõÑÎ°ú Í∞ÄÏ§ëÏπò Î≥ÄÌôîÍ∞Ä ÏûàÎäîÏßÄ ÌôïÏù∏ÌïòÎäî Ïö©ÎèÑÎ°ú Ïì∞Ïùº ÏòàÏ†ï
- matplotlibÏùÑ ÌôúÏö©Ìïú Í∞ÄÏ§ëÏπòÏóêÎåÄÌïú bar plotÏùÑ ÏÉùÏÑ±

```python
# Í∞ÄÏ§ëÏπò ÏãúÍ∞ÅÌôî Í∏∞Îä•
def weights_plot(weights):
    '''
    ÏûÖÎ†•Í∞í Ï†ïÎ≥¥
    weights : Í∞ÄÏ§ëÏπò (ndarray)
    '''
    # Í∞ÄÏ§ëÏπò ndarray -> pandas SeriesÎ°ú Î≥ÄÌôò(Ïù∏Îç±Ïã±Ïù¥ Îçî Ìé∏Ìï¥Ïßê)
    weights_series = pd.Series(np.reshape(weights,(-1)))
    
    # plot ÌÅ¨Í∏∞Î•º 6x6ÏúºÎ°ú ÏßÄÏ†ï
    plt.figure(figsize=(6,6))
    # barplot ÏÉùÏÑ±
    weights_bar = plt.bar(weights_series.index, weights_series, color = 'pink')
    
    # Í∞Å bar ÎßàÎã§ Í∞í textÍ∞Ä ÏúÑÏ™ΩÏóê ÌëúÏãúÎêòÎèÑÎ°ù Ìï¥Î≥¥Í∏∞
    for rect in weights_bar:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2.0, height, '%.3f' % height,
                 ha='center', va='bottom', size=8)
    # xÍ∞íÏùò ÏàúÏÑúÏùò ÏùòÎØ∏Í∞Ä ÏûàÎäî Í±¥ ÏïÑÎãàÎùºÏÑú xÏ∂ïÏùÄ ÌëúÏãúÎêòÏßÄ ÏïäÎèÑÎ°ù ÏÑ§Ï†ï
    plt.gca().axes.xaxis.set_visible(False)

    # barplot Ï∂úÎ†•
    plt.title("Barplot of weights")
    plt.show()
```

```python
# global Î≥ÄÏàòÏóê Ï†ÄÏû•ÎêòÏñ¥ÏûàÎäî weightsÏùò Î∂ÑÌè¨Î•º ÏãúÍ∞ÅÌôî Ìï¥Î≥¥Í∏∞
weights_plot(weights)
```

## ‚öôÔ∏è Í∏∞Îä• 3-3. Î∞∞ÏπòÎç∞Ïù¥ÌÑ∞ ÏñªÎäî Ìï®Ïàò : get_batch_data()
- ÏàúÏ†ÑÌåå Î∞è ÌïôÏäµÏùÑ ÏßÑÌñâÌï†Îïå Î∞∞ÏπòÎ•º Í≥†Î†§ÌïòÏó¨ Ïó∞ÏÇ∞ÏùÑ ÏßÑÌñâÌï† ÏòàÏ†ïÏù¥Í∏∞ ÎïåÎ¨∏Ïóê, Î∞∞ÏπòÎç∞Ïù¥ÌÑ∞Î•º ÏñªÎäî Ìï®ÏàòÎ•º Ï†ïÏùòÌï®
- Ïù∏Í≥µÏã†Í≤ΩÎßù classÏóêÏÑú forÎ¨∏ÏùÑ ÌÜµÌï¥ÏÑú Î∞∞ÏπòÎ≥ÑÎ°ú Î∞òÎ≥µÏùÑ ÏàòÌñâÌï†Îïå Ïì∞Ïùº ÏòàÏ†ï

```python
# Î∞∞ÏπòÎç∞Ïù¥ÌÑ∞ ÏñªÎäî Í∏∞Îä•
def get_batch_data(X, y, size, n):
    '''
    ÏûÖÎ†•Í∞íÏ†ïÎ≥¥
    X : ÎèÖÎ¶ΩÎ≥ÄÏàò(x1~x8)
    y : Ï¢ÖÏÜçÎ≥ÄÏàò(y)
    size : mini-batch ÌÅ¨Í∏∞
    n : 0 ~ (batchÏàò - 1) - for Î¨∏ÏùÑ ÌÜµÌï¥ ÏûÖÎ†•Î∞õÏùÑ Í∞íÏûÑ
    '''
    X_batch = X[size*n : size*(n+1)]
    y_batch = y[size*n : size*(n+1)]
    # Ï∂úÎ†• : Ïù∏Îç±Ïä§ batchÌÅ¨Í∏∞*nÎ∂ÄÌÑ∞ batchÌÅ¨Í∏∞ ÎßåÌÅºÏùò Îç∞Ïù¥ÌÑ∞
    return X_batch, y_batch
```

- Î∞∞ÏπòÎç∞Ïù¥ÌÑ∞ ÏñªÎäî Ìï®Ïàò ÎèôÏûë ÌÖåÏä§Ìä∏(Ï∂úÎ†•ÏóêÏÑ† Í∞ÑÎã®ÌïòÍ≤å yÏóê ÎåÄÌï¥ÏÑúÎßå ÌôïÏù∏Ìï¥Î≥¥ÎèÑÎ°ù ÌïòÍ≤†Ïùå)

```python
# batch sizeÎäî 4Î°ú ÏÑ§Ï†ï
batch_size = 4
# train data Ïùò Î∞∞Ïπò ÏàòÎ•º ÏñªÍ∏∞
batch_count = int(X_train.shape[0]) // batch_size
print(f"batch_size = {batch_size}, batch_count = {batch_count}")

# for Î¨∏ÏùÑ ÌÜµÌï¥ batchÎ•º Î∂ÑÎ¶¨
for batch in range(batch_count):
    X, y = get_batch_data(X_train, y_train, batch_size, batch)
    # Í∞ÑÎã®ÌïòÍ≤å ÌôïÏù∏ÌïòÍ∏∞ ÏúÑÌï¥ Ï¢ÖÏÜçÎ≥ÄÏàò(y)Îßå ÌôïÏù∏Ìï¥Î≥¥ÎèÑÎ°ù ÌïòÍ≤†Ïùå
    print(f"\nBatch{batch+1}")
    print(y)
```

## ‚öôÔ∏è Í∏∞Îä• 3-4. ÌôúÏÑ±ÌôîÌï®Ïàò : sigmoid()
- Ïù∏Í≥µÏã†Í≤ΩÎßùÏóêÏÑú ÌôúÏÑ±Ìôî Ìï®Ïàò Ïó≠Ìï†ÏùÑ ÌïòÎäî Ìï®ÏàòÎ•º Ï†ïÏùò, Î≥∏ Í≥ºÏ†úÏùò Ï°∞Í±¥Ïóê Îî∞Îùº sigmoidÎ•º ÌôúÏÑ±ÌôîÌï®ÏàòÎ°ú ÏÇ¨Ïö©
- sigmoid Í≥µÏãù
  - $sigmoid(y) = \frac{1}{1+e^{-y}}$

```python
# ÌôúÏÑ±ÌôîÌï®Ïàò : sigmoid (y : logits)
def sigmoid(y):
    '''
    ÏûÖÎ†•Í∞í Ï†ïÎ≥¥
    y : Í∞ÄÏ§ëÏπòÏôÄ Ìé∏Ìñ•ÏúºÎ°ú Ïó∞ÏÇ∞Îêú logits
    '''
    Y = 1 / (1 + np.exp(-y))
    
    # Ï∂úÎ†•Í∞í Y : ÌôúÏÑ±ÌôîÌï®ÏàòÎ•º Í±∞Ïπú ÏµúÏ¢Ö ÌôïÎ•†Í∞í
    return Y
```

- ÌôúÏÑ±Ìôî Ìï®Ïàò ÎèôÏûë ÌÖåÏä§Ìä∏

```python
# ÌôúÏÑ±Ìôî Ìï®ÏàòÏùò ÏûÖÎ†•Í∞íÏùÄ Í∞ÄÏ§ëÏπòÏôÄ Ìé∏Ìñ•ÏúºÎ°ú Ïó∞ÏÇ∞Îêú logits
example_y = np.matmul(X_train, weights) + bias

# sigmoid Ìï®ÏàòÎ•º ÌÜµÌï¥ ÌôúÏÑ±ÌôîÎêú Í∞í (y -> Y)
example_Y = sigmoid(example_y)
example_Y
```

## ‚öôÔ∏è Í∏∞Îä• 3-5. ÏÜêÏã§Ìï®Ïàò : sigmoid_crossentropy_logits()
- ÏàúÏ†ÑÌåå, Í≤ÄÏ¶ù, ÌïôÏäµÏóêÏÑú Ïì∞Ïùº ÏÜêÏã§Ìï®ÏàòÎ•º Ï†ïÏùòÌï®
  - ÏÜêÏã§Ìï®ÏàòÎäî Í≥ºÏ†ú Ï°∞Í±¥Ïóê Îî∞Îùº Ïù¥ÏßÑ ÍµêÏ∞® ÏóîÌä∏Î°úÌîº(binary cross entropy)Î°ú Ï†ïÏùò
- Í∞ÄÏ§ëÌï© Ïó∞ÏÇ∞ÏùÑ ÌÜµÌï¥ ÏñªÏùÄ logitsÎ°ú ÌôúÏÑ±ÌôîÌï®Ïàò Ï†ÅÏö©Í≥º ÏÜêÏã§Í∞í Í≥ÑÏÇ∞ÏùÄ ÌïúÎ≤àÏóê Ï≤òÎ¶¨Ìï®
  - Ïó¨Í∏∞ÏÑú logitsÎûÄ ÌôúÏÑ±ÌôîÌï®ÏàòÎ•º Í±∞ÏπòÏßÄ ÏïäÏùÄ Raw OutputÏùÑ ÏùòÎØ∏Ìï®
  - ÌôúÏÑ±Ìôî Ìï®ÏàòÎ•º Í±∞Ïπú ÌôïÎ•†Í∞íÏúºÎ°ú ÏÜêÏã§Í∞íÏùÑ Í≥ÑÏÇ∞ÌïòÎ©¥ ÏàúÏ†ÑÌååÍ≥ºÏ†ïÍπåÏßÑ Î¨∏Ï†úÍ∞Ä ÏóÜÏßÄÎßå, ÌïôÏäµÍ≥ºÏ†ïÏóêÏÑú ÌååÏù¥Ïç¨ÏóêÏÑú ÌóàÏö©ÌïòÎäî Ïó∞ÏÇ∞Ïùò Î≤îÏúÑÎ•º Ï¥àÍ≥ºÌï¥Î≤ÑÎ¶¨Í∏∞ ÎïåÎ¨∏Ïóê logitÏùÑ ÌÜµÌï¥ Íµ¨ÌïòÎäî Î∞©ÏãùÏùÑ Ï±ÑÌÉùÌïòÏòÄÏùå
  - Ïù¥Îü¨Ìïú Í≥ÑÏÇ∞ Î∞©ÏãùÏùÄ tensorflowÏóêÎèÑ Íµ¨ÌòÑÎêòÍ≥† Í∂åÏû•ÌïòÎäî Í≥ÑÏÇ∞Î∞©Î≤ïÏûÑ
    - tf.nn.sigmoid_cross_entropy_with_logits
    - https://www.tensorflow.org/api_docs/python/tf/nn/sigmoid_cross_entropy_with_logits
- sigmoid_cross_entropy_with_logits Í≥µÏãù Ïú†ÎèÑ
  - cross_entropy : $H(P,Q)= -Œ£P_i\space log\space Q_i$
    - ÏóîÌä∏Î°úÌîº $H$
    - labelÏùò Ï∂îÏ†ï ÌôïÎ•† $P$      
    - logit yÏóê ÎåÄÏùëÌïòÎäî ÏòàÏ∏°ÌôïÎ•† $Q$      
  - binary_cross_entropy : $H=-p_t\space log\space q_t-p_f\space log\space q_f$
    - Ï†ïÎãµ labelÍ∞í : $z$
    - Í∞ÄÏ§ëÌï© Ïó∞ÏÇ∞Ïóê ÏùòÌïú logit : $y$
    - ÏóîÌä∏Î°úÌîº $H$
    - labelÏùò Ï∂îÏ†ï ÌôïÎ•† $P$
      - binary : Ï∞∏ÏùºÌôïÎ•† $p_t=z$, Í±∞ÏßìÏùºÌôïÎ•† $p_f=1-p_t=1-z$
    - logit yÏóê ÎåÄÏùëÌïòÎäî ÏòàÏ∏°ÌôïÎ•† $Q$
      - binary : Ï∞∏ $q_t=sigmoid(y)$, Í±∞Ïßì $q_f=1-sigmoid(y)$
  - $H = - Œ£p_ilogq_i$
  - $H = -p_t\space log\space q_t-p_f\space log\space q_f$
  - $H = -z\space log(sigmoid(y))-(1-z)\space log(1-sigmoid(y))$
  - $H = -z\space log(\frac{1}{1+e^{-y}})-(1-z)\space log(1-\frac{1}{1+e^{-y}})$
  - $H = -z\space log(\frac{1}{1+e^{-y}})-(1-z)\space log(\frac{e^{-y}}{1+e^{-y}})$
  - $H = z\space log(1+e^{-y})-(1-z)(log(e^{-y})-log(1+e^{-y}))$
  - $H = z\space log(1+e^{-y})+(1-z)(y+log(1+e^{-y}))$
  - $H = z\space log(1+e^{-y})+y+log(1+e^{-y})-zy-z\space log(1+e^{-y})$
  - $H = y - zy + log(1+e^{-y})$
- sigmoid_crossentropy_logits ÏµúÏ¢Ö Í≥µÏãù
  - $E = y(1 - z) + log(1+e^{-y})$

```python
# ÏÜêÏã§ Ìï®Ïàò Í∏∞Îä• (from_logits=True, ÌôúÏÑ±Ìôî Í∏∞Îä•Í≥º ÏÜêÏã§Ìï®Ïàò Í∏∞Îä•ÏùÑ ÌïúÎ≤àÏóê Ïã§ÏãúÌï®)
def sigmoid_crossentropy_logits(z, y):
    '''
    ÏûÖÎ†•Í∞í Ï†ïÎ≥¥
    z : Ï†ïÎãµ data (Label data)
    y : Í∞ÄÏ§ëÏπòÏôÄ Ìé∏Ìñ•ÏúºÎ°ú Ïó∞ÏÇ∞Îêú logits
    '''
    E = y*(1-z) + np.log(1 + np.exp(-y))
    
    # Ï∂úÎ†•Í∞í E : ÏÜêÏã§Í∞í
    return E
```

- ÏÜêÏã§Ìï®Ïàò ÎèôÏûë ÌÖåÏä§Ìä∏

```python
# ÏÜêÏã§Ìï®ÏàòÎ°ú Í≥ÑÏÇ∞Îêú ÏÜêÏã§Í∞í (z,y -> E)
example_E = sigmoid_crossentropy_logits(y_train, example_y)
example_E
```

## ‚öôÔ∏è Í∏∞Îä• 3-6. ÏÜêÏã§Ìï®Ïàò Ìé∏ÎØ∏Î∂ÑÌï®Ïàò : sigmoid_crossentropy_logits_prime()
- Ïó≠Ï†ÑÌåå Í≥ºÏ†ïÏóêÏÑú Ïì∞Ïùº Ìé∏ÎØ∏Î∂Ñ Ìï®ÏàòÎ•º Ï†ïÏùòÌï®
  - ÏïûÏÑúÏÑú logitÏúºÎ°ú ÌôúÏÑ±ÌôîÏôÄ ÏÜêÏã§Í≥ÑÏÇ∞ÏùÑ ÌïúÎ≤àÏóê Ïó∞ÏÇ∞ÌñàÍ∏∞ ÎïåÎ¨∏Ïóê Ìé∏ÎØ∏Î∂Ñ ÎòêÌïú Ïù¥ÏôÄ ÎèôÏùºÌïú ÏûÖÎ†•Í∞íÏúºÎ°ú logitÏùÑ Ïù¥Ïö©Ìï®
  - dE/dY*dY/dy Îëê Í≥ÑÏÇ∞ Îã®Í≥ÑÎ•º dE/dyÎ°ú Î∞îÎ°ú Ï≤òÎ¶¨Ìï† Ïàò ÏûàÏúºÎØÄÎ°ú Ïó∞ÏÇ∞ ÏÜçÎèÑÎ©¥ÏóêÏÑúÎèÑ Ìõ®Ïî¨ Í¥úÏ∞ÆÏùÄ Î∞©ÏãùÏûÑ
- Ï¶â logit (y)Î°ú Ìé∏ÎØ∏Î∂ÑÌïòÏó¨ logitÏóêÎåÄÌïú ÏàúÍ∞Ñ Í∏∞Ïö∏Í∏∞Î•º Íµ¨Ìï† Ïàò ÏûàÍ≥†, Ïù¥Î•º X(=dy/dw)ÏôÄ ÌñâÎ†¨ Í≥±Ìï¥Ï£ºÎ©¥ Í∞ÄÏ§ëÏπòÎ•º ÏóÖÎç∞Ïù¥Ìä∏ Ìï† Ïàò ÏûàÏúºÎ©∞, Ìé∏Ìñ• ÎòêÌïú dE/dyÏùò Ìï©ÏúºÎ°úÏç® ÏóÖÎç∞Ïù¥Ìä∏ Ìï¥Ï§Ñ Ïàò ÏûàÍ≤å ÎêúÎã§
- dE/dy Ìé∏ÎØ∏Î∂Ñ Í≥µÏãù Ïú†ÎèÑ
  - $E = y(1 - z) + log(1+e^{-y})$
  - $\frac{dE}{dy}=(1-z) + \frac{(1+e^{-y})'}{1+e^{-y}}$
  - $\frac{dE}{dy}=(1-z) + \frac{-e^{-y}}{1+e^{-y}}$
  - $\frac{dE}{dy}=-z + 1 + \frac{-e^{-y}}{1+e^{-y}}$
  - $\frac{dE}{dy}=-z + \frac{1+e^{-y}-e^{-y}}{1+e^{-y}}$
  - $\frac{dE}{dy}=-z + \frac{1}{1+e^{-y}}$
  - $\frac{dE}{dy}=-z + sigmoid(y)$
- sigmoid_crossentropy_logits_prime ÏµúÏ¢Ö Í≥µÏãù
  - $\frac{dE}{dy}=-z + sigmoid(y)$

```python
# Ïó≠Ï†ÑÌååÎ•º ÏúÑÌïú ÏÜêÏã§Ìï®Ïàò Ìé∏ÎØ∏Î∂ÑÌï®Ïàò(logitÍ∞íÏúºÎ°ú Î∞îÎ°ú Ìé∏ÎØ∏Î∂ÑÌïòÏó¨ ÌôúÏÑ±ÌôîÌï®ÏàòÏùò Ìé∏ÎØ∏Î∂Ñ Îã®Í≥ÑÎ•º Í±¥ÎÑàÎúÄ)
def sigmoid_crossentropy_logits_prime(z, y):
    '''
    ÏûÖÎ†•Í∞í Ï†ïÎ≥¥
    z : Ï†ïÎãµ data (Label data)
    y : Í∞ÄÏ§ëÏπòÏôÄ Ìé∏Ìñ•ÏúºÎ°ú Ïó∞ÏÇ∞Îêú logits
    '''
    dE_dy = -z + sigmoid(y)
    
    # Ï∂úÎ†•Í∞í dE_dy : ÏÜêÏã§Í∞í EÎ•º logitÍ∞í yÎ°ú Ìé∏ÎØ∏Î∂ÑÌïòÏó¨ ÏñªÏñ¥ÎÇ∏ Í∏∞Ïö∏Í∏∞Í∞í
    return dE_dy
```

- ÏÜêÏã§Ìï®Ïàò Ìé∏ÎØ∏Î∂Ñ Ìï®Ïàò ÎèôÏûë ÌÖåÏä§Ìä∏

```python
# ÏÜêÏã§Í∞í EÎ•º logitÍ∞í yÎ°ú Ìé∏ÎØ∏Î∂ÑÌïòÏó¨ ÏñªÏñ¥ÎÇ∏ Í∏∞Ïö∏Í∏∞Í∞í (z,y -> dE/dy)
example_dE_dy = sigmoid_crossentropy_logits_prime(y_train, example_y)
example_dE_dy
```

## ‚öôÔ∏è Í∏∞Îä• 3-7. Ï†ïÌôïÎèÑ Ïó∞ÏÇ∞ Í∏∞Îä• : accuracy_score()
- numpyÏùò Í∏∞Îä•Îì§ÏùÑ ÌôúÏö©ÌïòÏó¨ Ï†ïÌôïÎèÑÎ•º Í≥ÑÏÇ∞ÌïòÎäî Ìï®ÏàòÎ•º Ï†ïÏùò
  - 0(False), 1(True)Î°ú Ïù¥Î§ÑÏßÑ BooleanÏùÑ Í≥ÑÏÇ∞ÌïòÎäî Î∞©ÏãùÏúºÎ°ú Ïó∞ÏÇ∞ Í≥ºÏ†ïÏùÑ Í∞ÑÎã®ÌïòÍ≤å ÏàòÌñâÌï®
- ÏûÖÎ†•ÏùÄ ÏÜêÏã§Ìï®ÏàòÏôÄ ÎßàÏ∞¨Í∞ÄÏßÄÎ°ú logitÍ≥º labelÍ∞íÏùÑ ÏûÖÎ†•Ìï®

```python
# Ï†ïÌôïÎèÑ Ïó∞ÏÇ∞ Í∏∞Îä•
def accuracy_score(z, y):
    '''
    ÏûÖÎ†•Í∞í Ï†ïÎ≥¥
    z : Ï†ïÎãµ data (Label data)
    y : Í∞ÄÏ§ëÏπòÏôÄ Ìé∏Ìñ•ÏúºÎ°ú Ïó∞ÏÇ∞Îêú logits
    '''    
    # pred : logitÏùò Î∂ÄÌò∏Ïóê Îî∞ÎùºÏÑú booleanÏúºÎ°ú Î∞òÌôò (0Ïù¥ÌïòÎäî False(0), 0Ï¥àÍ≥ºÎäî True(1))
    pred = np.greater(y, 0)
    # real : 0Í≥º 1Î°ú Ïù¥Î§ÑÏßÑ label dataÎ•º booleanÏúºÎ°ú Î∞òÌôò (0ÏùÄ False(0), 1ÏùÄ True(1))
    real = np.greater(z, 0.5)
    
    # correct : ÏòàÏ∏°Ïù¥ labelÍ≥º Í∞ôÏùÄ Í≤ΩÏö∞ True(1)Î°ú Î∞òÌôò
    correct = np.equal(pred, real)
    
    # accuracy : correctÎ•º ÌèâÍ∑†ÎÇ¥Î©¥ BooleanÏù¥ intÎ°ú Ïó∞ÏÇ∞ÎêòÏñ¥ Ï†ïÌôïÎèÑÍ∞Ä Î∞îÎ°ú Íµ¨Ìï¥ÏßÑÎã§
    accuracy = np.mean(correct)
    
    # Ï∂úÎ†•Í∞í : accuracy
    return accuracy
```

- Ï†ïÌôïÎèÑ Ïó∞ÏÇ∞ Í∏∞Îä• ÎèôÏûë ÌÖåÏä§Ìä∏

```python
accuracy_score(y_train, example_y)
```

```python
# ÏàúÏ†ÑÌååÍπåÏßÄÎßå ClassÎ°ú Íµ¨ÌòÑÌï¥Î≥¥Í∏∞(Ïù¥ÌõÑ Ï∂îÍ∞ÄÍ∏∞Îä•Îì§ÏùÄ ÏÉÅÏÜçÏùÑ ÌÜµÌï¥ Ï∂îÍ∞ÄÌï† ÏòàÏ†ï)
class NeuralNetwork:
    # Ï¥àÍ∏∞Ìï®Ïàò(Class ÏÑ†Ïñ∏Ïãú Ïã§ÌñâÎêòÎäî Ï¥àÍ∏∞ Ìï®Ïàò)
    def __init__(self, input_dim, output_dim):
        '''
        ÏûÖÎ†•Í∞í Ï†ïÎ≥¥
        input_dim : ÏûÖÎ†•Ï∏µ ÎÖ∏Îìú Ïàò
        output_dim : Ï∂úÎ†•Ï∏µ ÎÖ∏Îìú Ïàò
        '''
        # initialization_parameter Ìï®ÏàòÎ•º Ïã§ÌñâÏãúÏºú Ï¥àÍ∏∞ Í∞ÄÏ§ëÏπòÏôÄ Ìé∏Ìñ•ÏùÑ Ï†ÄÏû•
        self.w, self.b = initialization_parameter(input_dim, output_dim)
    
    # Í∞ÄÏ§ëÌï© Ìï®Ïàò (Í∞ÄÏ§ëÏπòÏôÄ Ìé∏Ìñ•ÏóêÎåÄÌïú Ïó∞ÏÇ∞ Í≤∞Í≥º Í∞í(logit)ÏùÑ ÏñªÏñ¥ÎÇ¥Îäî Í∏∞Îä•)
    def weight_sum(self, X):
        '''
        ÏûÖÎ†•Í∞í Ï†ïÎ≥¥
        X : ÎèÖÎ¶ΩÎ≥ÄÏàò(x1~x8)
        '''
        # y : XÏôÄ Í∞ÄÏ§ëÏπòÌñâÎ†¨(self.w)Ïùò ÌñâÎ†¨Í≥± + Ìé∏Ìñ•Í∞í
        y = np.matmul(X, self.w) + self.b
        return y
    
    # ÏàúÏ†ÑÌåå Í∏∞Îä•(ÌïôÏäµÍ≥ºÏ†ïÏù¥ ÏïÑÎãå Îã®Ïàú ÏàúÏ†ÑÌåå Í∏∞Îä•)
    def feed_forward(self, X_data, y_data, mb_size, verbose=1):
        '''
        ÏûÖÎ†•Í∞í Ï†ïÎ≥¥
        X_data : input Î≥ÄÏàò(x1~x8), X_train
        y_data : label Î≥ÄÏàò(y), y_train
        mb_size : Mini-Batch ÌÅ¨Í∏∞
        verbose : Í∏∞Î≥∏Í∞íÏùÄ 1, ÌõàÎ†® Í≤∞Í≥ºÎ•º Ï∂úÎ†• Ìï†ÏßÄ Ïó¨Î∂ÄÎ•º Í≤∞Ï†ïÌï®
            verbose = 0 : Ï∂úÎ†•ÏïàÌï®
            verbose = 1 : 1 epochÏóê ÎåÄÌïú ÌèâÍ∑† lossÏôÄ accuracyÏ∂úÎ†•
            verbose = 2 : batchÎ≥Ñ Îç∞Ïù¥ÌÑ∞ÍπåÏßÄ Ìï®Íªò Ï∂úÎ†•
        '''
        # Batch Í∞ØÏàòÎ•º Í≥ÑÏÇ∞
        batch_count = int(X_data.shape[0]) // mb_size
        # ÏÜêÏã§Í≥º Ï†ïÌôïÎèÑÎ•º Îã¥ÏùÑ Îπà Î¶¨Ïä§Ìä∏Î•º ÏÉùÏÑ±
        losses, accs = [], []
        
        # for Î¨∏ÏúºÎ°ú batch Ïó∞ÏÇ∞ÏùÑ Î∞òÎ≥µÌï®
        for batch in range(batch_count):
            # X(input), z(label)Î•º get_batch_data Ìï®ÏàòÎ°úÎ∂ÄÌÑ∞ ÏñªÏñ¥ÎÉÑ
            X, z = get_batch_data(X_data, y_data, size=mb_size, n=batch)
            # X(input) -> y(Í∞ÄÏ§ëÌï©) Ïó∞ÏÇ∞
            y = self.weight_sum(X=X)
            # Í∞Å Í∞íÎì§Ïùò ÏÜêÏã§Í∞í(E) Ïó∞ÏÇ∞
            E = sigmoid_crossentropy_logits(z=z, y=y)
            # loss : ÎØ∏ÎãàÎ∞∞ÏπòÏùò ÌèâÍ∑† ÏÜêÏã§Í∞í
            loss = np.mean(E)
            # accuracy : ÎØ∏ÎãàÎ∞∞ÏπòÏùò Ï†ïÌôïÎèÑÍ∞í
            accuracy = accuracy_score(z=z, y=y)
            
            # ÏÜêÏã§Í∞íÍ≥º Ï†ïÌôïÎèÑÎ•º Î¶¨Ïä§Ìä∏Ïóê appendÌïòÍ∏∞(ÏÜåÏàòÏ†ê ÏÖãÏß∏ÏûêÎ¶¨Î°ú Î∞òÏò¨Î¶º)
            losses.append(round(loss, 3))
            accs.append(round(accuracy, 3))
        
        # verbose Í∏∞Îä•(Í∏∞Î≥∏Í∞íÏùÄ 1)
        if verbose in [1,2]:
            # ÌèâÍ∑† lossÏôÄ ÌèâÍ∑† accuracyÎ•º Ï∂úÎ†•(ÏÜåÏàòÏ†ê ÏÖãÏß∏ÏûêÎ¶¨Î°ú Î∞òÏò¨Î¶º)
            print("[Epoch 1] TrainData - Loss = {:.3f}, Accuracy = {:.3f}"
                  .format(np.mean(losses), np.mean(accs)))
            if verbose == 2:
                # Í∞Å Î∞∞ÏπòÎ≥Ñ lossÏôÄ accuracyÎ•º ÌôïÏù∏ÌïòÍ∏∞ ÏúÑÌïú Ï∂úÎ†•(verbose==2Î°ú ÏÑ§Ï†ïÌï¥Ïïº Î≥¥ÏûÑ)
                print(f"\tBatch Size : {mb_size}\n\tMini-Batches : {batch_count}\
                    \n\tLoss : {losses}\n\tAccuracy : {accs}")
        elif verbose == 0:
            # verbose = 0 ÏúºÎ°ú Ï£ºÏñ¥ÏßÄÎ©¥ Ï∂úÎ†•ÏïàÌïòÍ≥† passÌï®
            pass
    
    # ÏòàÏ∏° Í∏∞Îä•(prediction, ndarrayÎ°ú Î∞òÌôòÌï®)
    def predict(self, X_data, threshold=0.5):
        '''
        ÏûÖÎ†•Í∞í Ï†ïÎ≥¥
        X_data : input Î≥ÄÏàò(x1~x8)
        threshold : Í∏∞Î≥∏Í∞íÏùÄ 0.5, 0Í≥º 1ÏùÑ Íµ¨Î∂ÑÌï† ÏûÑÍ≥ÑÏπò(Í∏∞Ï§ÄÏÑ†)
        '''
        # Í∞ÄÏ§ëÌï© Ïó∞ÏÇ∞
        y = self.weight_sum(X_data)
        # ÌôúÏÑ±Ìôî Ìï®ÏàòÎ•º Í±∞Ï≥ê ÌôïÎ•†Í∞íÏùÑ ÏñªÏùå
        Y = sigmoid(y)
        # ThresholdÎ≥¥Îã§ ÏûëÏúºÎ©¥ 0ÏúºÎ°ú ÏòàÏ∏°, ÌÅ¨Í±∞ÎÇò Í∞ôÏúºÎ©¥ 1Î°ú ÏòàÏ∏°
        pred = np.where(Y < threshold, 0, 1)
        
        # Ï∂úÎ†• : ÏòàÏ∏°Í∞í(ndarray)
        return pred
    
    # Í≤ÄÏ¶ù Í∏∞Îä•(evaluation, ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌïú ÏàúÏ†ÑÌåå Í≤∞Í≥ºÎ•º Î∞îÎ°ú Ï∂úÎ†•)
    def evaluate(self, X_data, y_data):
        '''
        ÏûÖÎ†•Í∞í Ï†ïÎ≥¥
        X_data : input Î≥ÄÏàò(x1~x8), X_test
        y_data : label Î≥ÄÏàò(y), y_test
        '''
        # X(input), z(label) Ï†ïÏùò
        X, z = X_data, y_data
        # Í∞ÄÏ§ëÌï©(logits) Ïó∞ÏÇ∞
        y = self.weight_sum(X_data)
        
        # ÌèâÍ∑† lossÍ∞í Ïó∞ÏÇ∞
        loss = np.mean(sigmoid_crossentropy_logits(z=z, y=y))
        # ÌèâÍ∑† accuracy Ïó∞ÏÇ∞
        accuracy = accuracy_score(z=z, y=y)
        
        # Í≤ÄÏ¶ù Í≤∞Í≥º Ï∂úÎ†•ÌïòÍ∏∞ (ÏÜåÏàòÏ†ê ÏÖãÏß∏ÏûêÎ¶¨Î°ú Î∞òÏò¨Î¶º)
        print("[Evaluation] TestData - Loss = {:.3f}, Accuracy = {:.3f}".format(loss, accuracy))
    
    # Í∞ÄÏ§ëÏπò ÏãúÍ∞ÅÌôî Ìï®Ïàò(Í∞ÄÏ§ëÏπò Î≥ÄÌôîÎ•º ÌôïÏù∏Ìï¥Î≥¥Í∏∞ ÏúÑÌï¥ classÎÇ¥Î∂ÄÏóê Ìè¨Ìï®ÏãúÏº∞Ïùå)
    def weights_plot(self):
        # self.wÎ°ú Ï†ÄÏû•ÎêòÏñ¥ÏûàÎäî Í∞ÄÏ§ëÏπòÎ•º bar plotÏúºÎ°ú ÏãúÍ∞ÅÌôîÌï®
        weights_plot(self.w)
```

## üß† Ïù∏Í≥µÏã†Í≤ΩÎßù(ANN) ÌÅ¥ÎûòÏä§(Class) Ï∂îÍ∞ÄÍµ¨ÌòÑ : NeuralNetwork_additional() - Ïó≠Ï†ÑÌåå, ÌïôÏäµ Í∏∞Îä•ÏùÑ Ï∂îÍ∞Ä Íµ¨ÌòÑ
- Ï∂îÍ∞ÄÏ†ÅÏúºÎ°ú Í∏∞Ï°¥ ÏàúÏ†ÑÌåå Ïù∏Í≥µÏã†Í≤ΩÎßùÏùÑ ÏÉÅÏÜçÎ∞õÏïÑÏÑú Ïó≠Ï†ÑÌåå, ÌïôÏäµ Í∏∞Îä•ÏùÑ Ï∂îÍ∞ÄÌï¥Î≥∏Îã§
- ÌïôÏäµ Ïù∏Í≥µÏã†Í≤ΩÎßù ÌÅ¥ÎûòÏä§(Class) Íµ¨ÏÑ± ÏöîÏÜå
  - Ï¥àÍ∏∞Ìï®Ïàò(`__init__`)
    - ÏàúÏ†ÑÌåå Ìï®ÏàòÏùò Ï¥àÍ∏∞Ìï®ÏàòÎ•º super()Î°ú Í∞ÄÏ†∏ÏôÄÏÑú Í∑∏ÎåÄÎ°ú Ïã§ÌñâÌï®
    - Ï¥àÍ∏∞ Í∞ÄÏ§ëÏπòÏôÄ Ìé∏Ìñ•Ïù¥ self.w, self.b Ïóê Ï†ÄÏû•ÎêúÎã§
  - Ïó≠Ï†ÑÌåå Í∏∞Îä• : feed_backward(X,z,y)
    - ÏûÖÎ†•Ï∏µ(X), label(z), logit(y)ÏùÑ ÏûÖÎ†•Î∞õÍ≥† Í∞ÄÏ§ëÏπòÏôÄ Ìé∏Ìñ•Ïùò ÏàúÍ∞ÑÍ∏∞Ïö∏Í∏∞Î•º Íµ¨ÌïúÎã§
    - dE/dy Îäî sigmoid_crossentropy_logits_prime Ìï®ÏàòÎ°ú Íµ¨Ìï®
    - dE/dw Îäî XÏùò Ïó≠ÌñâÎ†¨(dy/dw)Í≥º dE/dyÏùò ÌñâÎ†¨Í≥±ÏúºÎ°ú Íµ¨Ìï®
    - dE/db Îäî dE/dyÏùò Ïó¥Î∞©Ìñ•(axis=0) Ìï©Í≥ÑÎ°ú Íµ¨Ìï®
    - dE/dwÏôÄ dE/dbÎäî ÌÅ¥ÎûòÏä§ÏóêÏÑú Ï†ÑÏó≠Ï†ÅÏúºÎ°ú ÏÇ¨Ïö©ÎêòÎèÑÎ°ù self.dE_dw, self.dE_dbÎ°ú Ï†ÄÏû•ÌïúÎã§
  - ÌïôÏäµÍ∏∞Îä• : training(X, y, batch_size, epoch, learning_rate, verbose)
    - ÏßÄÏ†ïÌïú epoch ÏàòÎßåÌÅº ÏàúÏ†ÑÌåå, Ïó≠Ï†ÑÌåå, Í∞ÄÏ§ëÏπòÌé∏Ìñ• ÏóÖÎç∞Ïù¥Ìä∏Î•º ÏàòÌñâÌïúÎã§
    - Î∞∞ÏπòÎ•º Í≥†Î†§Ìï¥ ÎØ∏ÎãàÎ∞∞ÏπòÎ°ú Ïó∞ÏÇ∞ÏùÑ ÏßÑÌñâÌïòÍ≥†, ÏÜêÏã§Í∞íÍ≥º Ï†ïÌôïÎèÑÎ•º Íµ¨ÌïúÎã§
    - XÍ∞í(ÏûÖÎ†•), zÍ∞í(label), yÍ∞í(logitÍ∞í)ÏúºÎ°ú Ïó≠Ï†ÑÌååÎ•º ÏßÑÌñâÌïòÍ≥† Í∞ÄÏ§ëÏπòÏôÄ Ìé∏Ìñ•ÏùÑ ÌïôÏäµÎ•†Í≥º Í≥±Ìï¥ÏÑú ÏóÖÎç∞Ïù¥Ìä∏Î•º ÏßÑÌñâÌïúÎã§
    - Ìïú epochÍ∞Ä ÎÅùÎÇòÎ©¥ ÌèâÍ∑† ÏÜêÏã§Í∞íÍ≥º Ï†ïÌôïÎèÑÎ•º history ÎîïÏÖîÎÑàÎ¶¨Ïóê Ï†ÄÏû•Ìï¥ÎëêÍ≥† Îã§Ïùå epochÎ°ú Î∞òÎ≥µÏùÑ Ïã§ÏãúÌïúÎã§
    - verboseÏùÑ ÌÜµÌï¥ Ï∂úÎ†•ÏùÑ ÌïòÏßÄ ÏïäÏùÑÏßÄ, Î∞∞ÏπòÎ≥Ñ Îç∞Ïù¥ÌÑ∞ÍπåÏßÄ Î™®Îëê ÌôïÏù∏Ìï†ÏßÄ Í≤∞Ï†ïÌïúÎã§

image.png

```python
# Ïù¥Ï†ÑÏóê Ï†ïÏùòÌïú NeuralNetwork classÎ•º ÏÉÅÏÜçÎ∞õÍ≥† Ï∂îÍ∞ÄÏ†ÅÏúºÎ°ú Ïó≠Ï†ÑÌååÏôÄ ÌïôÏäµÍ∏∞Îä•ÏùÑ Íµ¨ÌòÑÌïòÍ∏∞
class NeuralNetwork_additional(NeuralNetwork):
    # Ï¥àÍ∏∞Ìï®Ïàò(Class ÏÑ†Ïñ∏Ïãú Ïã§ÌñâÎêòÎäî Ï¥àÍ∏∞Ìï®Ïàò : ÏÉÅÏÜçÎ∞õÏïÑÏÑú Í∑∏ÎåÄÎ°ú ÏÇ¨Ïö©Ìï®)
    def __init__(self, input_dim, output_dim):
        '''
        ÏûÖÎ†•Í∞í Ï†ïÎ≥¥
        input_dim : ÏûÖÎ†•Ï∏µ ÎÖ∏Îìú Ïàò
        output_dim : Ï∂úÎ†•Ï∏µ ÎÖ∏Îìú Ïàò
        '''
        # super()Î•º ÌÜµÌï¥ÏÑú self.w, self.b Î≥ÄÏàòÏ†ïÎ≥¥Î•º Í∑∏ÎåÄÎ°ú Î∞õÏïÑÏôÄÏÑú Ïã§ÌñâÌï®
        super().__init__(input_dim, output_dim)
    
    # Ïó≠Ï†ÑÌåå Í∏∞Îä• (ÌôúÏÑ±Ìôî Îã®Í≥ÑÎèÑ Ìï®Íªò Í≥ÑÏÇ∞ÌïòÎäî sigmoid_crossentropy_with_logits Î∞©Ïãù ÏÇ¨Ïö©)
    def feed_backward(self, X, z, y):
        '''
        ÏûÖÎ†•Í∞í Ï†ïÎ≥¥
        X : input Î≥ÄÏàò(x1~x8), X_train
        z : label Î≥ÄÏàò(y), y_train
        y : Í∞ÄÏ§ëÌï© Ïó∞ÏÇ∞ÏùÑ ÌÜµÌï¥ ÏñªÏùÄ logits
        '''
        # dE/dy Ïó∞ÏÇ∞ (sigmoid_crossentropy_logits_prime)
        dE_dy = sigmoid_crossentropy_logits_prime(z=z, y=y)
        # dy/dw = XÏùò Ïó≠ÌñâÎ†¨(transpose)
        dy_dw = X.T
        
        # dE/dw Í∞ÄÏ§ëÏπò Í∏∞Ïö∏Í∏∞Í∞í Ïó∞ÏÇ∞ (matmulÏùÑ ÌÜµÌïú ÌñâÎ†¨Í≥±)
        self.dE_dw = np.matmul(dy_dw, dE_dy)
        # dE/db Ìé∏Ìñ• Í∏∞Ïö∏Í∏∞Í∞í Ïó∞ÏÇ∞ (dE/dyÏùò Ìï©), biasÏôÄ ÌòïÌÉú ÎßûÏ∂îÍ∏∞ÏúÑÌï¥ axis=0ÏúºÎ°ú ÏÑ§Ï†ï(Ïó¥ Î∞©Ìñ•)
        self.dE_db = np.sum(dE_dy, axis=0) # (4x1) -> (1x1)
    
    # ÌïôÏäµ Í∏∞Îä• (EpochÎ≥Ñ batchÎ•º Í≥†Î†§Ìïú ÌïôÏäµÏùÑ ÏßÑÌñâ)
    def training(self, X_data, y_data, mb_size, epochs, learning_rate=1, verbose=1):
        '''
        ÏûÖÎ†•Í∞í Ï†ïÎ≥¥
        X_data : input Î≥ÄÏàò(x1~x8), X_train
        y_data : label Î≥ÄÏàò(y), y_train
        mb_size : Mini-Batch ÌÅ¨Í∏∞
        epochs : Epoch Ïàò
        learning_rate : Í∏∞Î≥∏Í∞íÏùÄ 1Î°ú ÏÑ§Ï†ïÌï®, ÌïôÏäµÎ•†
        verbose : Í∏∞Î≥∏Í∞íÏùÄ 1, ÌõàÎ†® Í≤∞Í≥ºÎ•º Ï∂úÎ†• Ìï†ÏßÄ Ïó¨Î∂ÄÎ•º Í≤∞Ï†ïÌï®
            verbose = 0 : Ï∂úÎ†•ÏïàÌï®
            verbose = 1 : 1 epochÏóê ÎåÄÌïú ÌèâÍ∑† lossÏôÄ accuracyÏ∂úÎ†•
            verbose = 2 : batchÎ≥Ñ Îç∞Ïù¥ÌÑ∞ÍπåÏßÄ Ìï®Íªò Ï∂úÎ†•
        '''
        # learning_rateÎ•º lrÎ°ú Ï†ÄÏû•Ìï®
        lr = learning_rate
        # Batch Í∞ØÏàòÎ•º Í≥ÑÏÇ∞
        batch_count = int(X_data.shape[0]) // mb_size
        # EpochÎ≥Ñ ÏÜêÏã§Í≥º Ï†ïÌôïÎèÑ Î¶¨Ïä§Ìä∏Î•º Îã¥ÏùÑ history ÎîïÏÖîÎÑàÎ¶¨Î•º ÏßÄÏ†ï
        self.history = {'Loss': [], 'Accuracy': []}
        
        # ÏßÄÏ†ïÌïú epoch ÏàòÏóê Îî∞ÎùºÏÑú Î∞òÎ≥µÏùÑ ÏßÑÌñâ
        for epoch in range(epochs):
            # ÏÜêÏã§Í≥º Ï†ïÌôïÎèÑÎ•º Îã¥ÏùÑ Îπà Î¶¨Ïä§Ìä∏ ÏßÄÏ†ï(epochÎßàÎã§ Ï¥àÍ∏∞ÌôîÎê®)
            losses, accs = [], []
            
            # batch Îã®ÏúÑÎ°ú Î∞òÎ≥µÏùÑ ÏßÑÌñâÌï®
            for batch in range(batch_count):
                # X(input), z(label)Î•º get_batch_data Ìï®ÏàòÎ°úÎ∂ÄÌÑ∞ ÏñªÏñ¥ÎÉÑ
                X, z = get_batch_data(X_data, y_data, size=mb_size, n=batch)
                # X(input) -> y(Í∞ÄÏ§ëÌï©) Ïó∞ÏÇ∞
                y = self.weight_sum(X=X)
                
                # Í∞Å Í∞íÎì§Ïùò ÏÜêÏã§Í∞í(E) Ïó∞ÏÇ∞
                E = sigmoid_crossentropy_logits(z=z, y=y)
                # loss : ÎØ∏ÎãàÎ∞∞ÏπòÏùò ÌèâÍ∑† ÏÜêÏã§Í∞í
                loss = np.mean(E)
                # accuracy : ÎØ∏ÎãàÎ∞∞ÏπòÏùò Ï†ïÌôïÎèÑÍ∞í
                accuracy = accuracy_score(z=z, y=y)
                
                # ÏÜêÏã§Í∞íÍ≥º Ï†ïÌôïÎèÑÍ∞íÏùÑ append (ÏÜåÏàòÏ†ê ÏÖãÏß∏ÏûêÎ¶¨Î°ú Î∞òÏò¨Î¶º)
                losses.append(round(loss, 3))
                accs.append(round(accuracy, 3))
                
                # Ïó≠Ï†ÑÌåå Ïã§Ïãú(self.dE_dw, self.dE_dbÍ∞Ä ÏóÖÎç∞Ïù¥Ìä∏ Îê®)
                self.feed_backward(X=X, z=z, y=y)
                # Í∞ÄÏ§ëÏπò ÏóÖÎç∞Ïù¥Ìä∏ Ïã§Ïãú
                self.w -= lr*self.dE_dw
                # Ìé∏Ìñ• ÏóÖÎç∞Ïù¥Ìä∏ Ïã§Ïãú
                self.b -= lr*self.dE_db
            
            # epochÎãπ ÌèâÍ∑† lossÏôÄ ÌèâÍ∑† accuracyÎ•º ÎîïÏÖîÎÑàÎ¶¨Ïùò Î¶¨Ïä§Ìä∏Ïóê append (ÏÜåÏàòÏ†ê ÏÖãÏß∏ÏûêÎ¶¨Î°ú Î∞òÏò¨Î¶º)
            self.history['Loss'].append(round(np.mean(losses), 3))
            self.history['Accuracy'].append(round(np.mean(accs), 3))
            
            # verbose Í∏∞Îä•(Í∏∞Î≥∏Í∞íÏùÄ 1)
            if verbose in [1,2]:
                # ÌèâÍ∑† lossÏôÄ ÌèâÍ∑† accuracyÎ•º Ï∂úÎ†•(ÏÜåÏàòÏ†ê ÏÖãÏß∏ÏûêÎ¶¨Î°ú Î∞òÏò¨Î¶º)
                print("\n[Epoch {}/{}] TrainData - Loss = {:.3f}, Accuracy = {:.3f}"
                      .format(epoch+1, epochs, np.mean(losses), np.mean(accs)))
                if verbose == 2:
                    # Í∞Å Î∞∞ÏπòÎ≥Ñ lossÏôÄ accuracyÎ•º ÌôïÏù∏ÌïòÍ∏∞ ÏúÑÌïú Ï∂úÎ†•(verbose==2Î°ú ÏÑ§Ï†ïÌï¥Ïïº Î≥¥ÏûÑ)
                    print(f"\tBatch Size : {mb_size}\n\tMini-Batchs : {batch_count}\
                        \n\tLoss : {losses}\n\tAccurracy : {accs}")
            elif verbose == 0:
                # verbose = 0 ÏúºÎ°ú Ï£ºÏñ¥ÏßÄÎ©¥ Ï∂úÎ†•ÏïàÌïòÍ≥† passÌï®
                pass
```

## üìà Í∏∞Îä•3-8. ÌïôÏäµ Í≥°ÏÑ† ÏãúÍ∞ÅÌôî Í∏∞Îä• : plot_history()
- Class ÎÇ¥Î∂ÄÏóêÏÑú Ï†ÄÏû•ÌñàÏóàÎçò self.history ÎîïÏÖîÎÑàÎ¶¨Î•º ÏãúÍ∞ÅÌôîÌïòÏó¨ epochÏóê Îî∞Î•∏ ÏÜêÏã§Ìï®Ïàò Î∞è Ï†ïÌôïÎèÑÏùò Î≥ÄÌôîÎ•º ÏãúÍ∞ÅÌôîÌïúÎã§
- matplotlibÏùÑ ÌôúÏö©ÌïòÏó¨ ÏãúÍ∞ÅÌôî ÏßÑÌñâ

```python
# ÌïôÏäµ Í≥°ÏÑ† ÏãúÍ∞ÅÌôî Í∏∞Îä•
def plot_history(NN, key, show_value=False, show_xticks=False):
    '''
    ÏûÖÎ†•Í∞í Ï†ïÎ≥¥
    NN : NeuralNetwork Ïù∏Ïä§ÌÑ¥Ïä§
    key : historyÎ≥ÄÏàòÏùò keyÍ∞í, 'Loss' or 'Accuracy'
    show_value : Í∑∏ÎûòÌîÑÏóê Í∞íÏùÑ ÌëúÏãúÌï†ÏßÄ Ïó¨Î∂Ä (Í∏∞Î≥∏Í∞íÏùÄ FalseÎ°ú ÏÑ§Ï†ï)
    show_xticks : Í∑∏ÎûòÌîÑÏóê xÏ∂ï Í∞íÏùÑ ÌëúÏãúÌï†ÏßÄ Ïó¨Î∂Ä (Í∏∞Î≥∏Í∞íÏùÄ FalseÎ°ú ÏÑ§Ï†ï)
    '''
    # Í∑∏ÎûòÌîÑ y Í∞í : keyÏóê Ìï¥ÎãπÌïòÎäî Î¶¨Ïä§Ìä∏
    y = NN.history[key]
    # Í∑∏ÎûòÌîÑ x Í∞í : epoch(1 ~ yÍ∏∏Ïù¥)
    x = np.arange(len(y))+1
    
    # Í∑∏ÎûòÌîÑ ÌÅ¨Í∏∞Î•º 6x6ÏúºÎ°ú ÏÑ§Ï†ï
    plt.figure(figsize=(6,6))
    # ÏÑ†Í∑∏ÎûòÌîÑ plot ('o-': Ï†êÌëúÏãú)
    plt.plot(y, 'o-', color='orange')
    
    if show_xticks == True:
        # xÏ∂ï ÎààÍ∏à ÏÑ§Ï†ï
        plt.xticks(np.arange(len(y)), labels=x)
    elif show_xticks == False:
        plt.xticks([])
    
    # yÏ∂ï Î≤îÏúÑ ÏÑ§Ï†ï
    ylim = [0 , np.trunc(max(y)+1)]
    plt.ylim(ylim)
    
    # Ï∂ï Ïù¥Î¶Ñ, Ï†úÎ™© ÏÑ§Ï†ï
    plt.xlabel("Epoch")
    plt.ylabel(f"{key}")
    plt.title(f'History of "{key}" in Neural Network')
    
    if show_value == True:
        # Í∑∏ÎûòÌîÑÏóê Îç∞Ïù¥ÌÑ∞ Í∞íÏùÑ ÌëúÏãúÌïòÍ∏∞
        for i in range(len(x)):
            height = y[i]
            plt.text(x[i]-1, height + ylim[1]/100, '%.3f' % height,
                     ha='center', va='bottom', size = 8)
    
    plt.show()
```

## üíΩ Í∏∞Îä• 4-2. Ï†ÑÏ≤¥ Í∏∞Îä• ÎèôÏûë Í∏∞Îä•(ÌïôÏäµ, ÏãúÍ∞ÅÌôî, ÏòàÏ∏°, Í≤ÄÏ¶ù) : main_additional()

```python
def main_additional():
    # 1-1. Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞ Í∏∞Îä•
    df = data_load(filepath=csv_dir)
    
    # 2-1. Îç∞Ïù¥ÌÑ∞ Îí§ÏÑûÍ∏∞ Í∏∞Îä•
    df_sf = data_shuffle(df)
    
    # 2-2. Îç∞Ïù¥ÌÑ∞ ÌëúÏ§ÄÌôî Í∏∞Îä•
    df_sc = standard_scaler(df_sf, output_dim=1)
    
    # 2-3. ÌïôÏäµ Î∞è ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨ Í∏∞Îä•
    train, test = train_test_split(df_sc, train_ratio=0.8, batch_size=4)
    
    # 2-4. ÎèÖÎ¶ΩÎ≥ÄÏàò Î∞è Ï¢ÖÏÜçÎ≥ÄÏàò Î∂ÑÎ¶¨
    X_train, y_train = X_y_split(train, output_dim=1)
    X_test, y_test = X_y_split(test, output_dim=1)
    
    # 3-1. Ïù∏Í≥µÏã†Í≤ΩÎßù(ANN) - Ïó≠Ï†ÑÌåå, ÌïôÏäµ Ï∂îÍ∞Ä Íµ¨ÌòÑ
    NN = NeuralNetwork_additional(input_dim=8, output_dim=1)
    NN.training(X_train, y_train, mb_size=4, epochs=5, learning_rate=1, verbose=2)
    
    # 3-2. ÌïôÏäµÍ≥°ÏÑ† ÏãúÍ∞ÅÌôî
    plot_history(NN, key='Accuracy', show_value=True, show_xticks=True)
    
    # 3-3. TestData ÏòàÏ∏°
    print(f"[Label_data] TestData\n{y_test}\n")
    y_pred = NN.predict(X_test)
    print(f"[Prediction] TestData\n{y_pred}\n")
    
    # 3-4. TestData Í≤ÄÏ¶ù
    NN.evaluate(X_test, y_test)
```
